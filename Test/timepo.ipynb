# Rellenar Saltos

```python
import pandas as pd

# Cargar CSV
df = pd.read_csv("lecturas_simuladas5.csv", parse_dates=["TimestampLocal"])

# Establecer índice de tiempo
df = df.set_index("TimestampLocal")

# Reindexar con frecuencia de 1 segundo
df_resample = df.resample("1S").mean()

# Interpolación spline cúbica (orden 3)
df_interp = df_resample.interpolate(method="time", order=3)

# Resetear índice si lo necesitas
df_interp = df_interp.reset_index()

# Guardar a un nuevo CSV
df_interp.to_csv("datos_interpolados1.csv", index=False)
```

```python
import pandas as pd
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# Cargar CSVs
df_modelado = pd.read_csv("niveles_tanques_cm1.csv")   # Contiene NivelTanque1, NivelTanque2
df_medido   = pd.read_csv("niveles_tanques_pool_cm1.csv")     # Contiene Nivel1, Nivel2

# Asegúrate de que estén alineados. Si tienen timestamps, puedes alinear por ellos.
# Por ejemplo, si tienen 'Timestamp':
# df_modelado = df_modelado.set_index('Timestamp')
# df_medido = df_medido.set_index('Timestamp')
# df = df_modelado.join(df_medido, how='inner')

# Extraer columnas
nivel1_modelado = df_modelado["NivelTanque1"]
nivel2_modelado = df_modelado["NivelTanque2"]
nivel1_medido   = df_medido["Nivel1"]
nivel2_medido   = df_medido["Nivel2"]

# Cálculo de errores
mae1 = mean_absolute_error(nivel1_medido, nivel1_modelado)
rmse1 = np.sqrt(mean_squared_error(nivel1_medido, nivel1_modelado))

mae2 = mean_absolute_error(nivel2_medido, nivel2_modelado)
rmse2 = np.sqrt(mean_squared_error(nivel2_medido, nivel2_modelado))

# Mostrar resultados
print("Errores para Nivel 1:")
print(f"  MAE:  {mae1:.4f}")
print(f"  RMSE: {rmse1:.4f}\n")

print("Errores para Nivel 2:")
print(f"  MAE:  {mae2:.4f}")
print(f"  RMSE: {rmse2:.4f}")
```

```python
# MAPE (error porcentual)
mape1 = np.mean(np.abs((nivel1_medido - nivel1_modelado) / nivel1_medido)) * 100
mape2 = np.mean(np.abs((nivel2_medido - nivel2_modelado) / nivel2_medido)) * 100

# NRMSE (error relativo normalizado)
nrmse1 = rmse1 / (nivel1_medido.max() - nivel1_medido.min())
nrmse2 = rmse2 / (nivel2_medido.max() - nivel2_medido.min())

print(f"MAPE Nivel 1: {mape1:.2f}%")
print(f"NRMSE Nivel 1: {nrmse1:.4f}\n")

print(f"MAPE Nivel 2: {mape2:.2f}%")
print(f"NRMSE Nivel 2: {nrmse2:.4f}")
```
